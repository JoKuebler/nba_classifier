```python
import pandas as pd

season_data_21 = pd.read_csv('data/box_scores/season_box_ext_20202021.csv', converters={'GAME_ID': str})
season_data_20 = pd.read_csv(
    'data/box_scores/season_box_ext_20192020.csv', converters={'GAME_ID': str})
season_data_19 = pd.read_csv(
    'data/box_scores/season_box_ext_20182019.csv', converters={'GAME_ID': str})

# Concat all season data
season_all = pd.concat([season_data_21, season_data_20, season_data_19])

# Move GAME_ID to the front
season_all = season_all[["GAME_ID", "PTS_PAINT", "PTS_2ND_CHANCE", "PTS_FB", "LARGEST_LEAD", "LEAD_CHANGES", "TIMES_TIED", "TEAM_REBOUNDS", "PTS_OFF_TO", "PTS_QTR1",
                           "PTS_QTR2", "PTS_QTR3", "PTS_QTR4", "PTS_OT1", "PTS_OT2", "PTS_OT3", "PTS_OT4", "PTS_OT5", "PTS_OT6", "PTS_OT7", "PTS_OT8", "PTS_OT9", "PTS_OT10", "PTS", "W_PCT", "FG_PCT", "FG3_PCT", "FT_PCT", "OREB", "AST", "STL", "BLK", "TOV"]]

# Convert all columns to float expect GAME_ID
season_all = season_all.astype(
    {col: float for col in season_all.columns if col != 'GAME_ID'})

season_all = pd.concat([season_all, season_all.add_suffix('_1')], axis=1)

# # Create a list to store the combined rows
# combined_all = pd.DataFrame()
# # Iterate through the DataFrame, skipping every second row
# for i in range(0, len(season_all), 2):
#     tmp_df = pd.concat([season_all.iloc[i], season_all.iloc[i + 1]], axis=1)
#     combined_all = pd.concat([combined_all, tmp_df])

# Save to csv to data/training folder 
combined_all.to_csv('data/training/train_all.csv', index=False)
```

```python
# Find comebacks
df_comeback = pd.DataFrame(columns=season_all.columns)
# Iterate over all rows pairwise based on GAME_ID
for i in range(0, len(season_all), 2):
    # Extract two rows at a time
    game_rows = season_all.iloc[i:i+2, :]

    # Get the row index of the max value in LARGEST_LEAD column
    # Largest Lead is not the winner
    if game_rows['LARGEST_LEAD'].idxmax() != game_rows['PTS'].idxmax():
        # if difference of largest lead column is larger then 10
        if game_rows['LARGEST_LEAD'].max() - game_rows['LARGEST_LEAD'].min() > 10:
            # Add the two rows to the comeback dataframe
            df_comeback = pd.concat([df_comeback, game_rows])

print(len(df_comeback))
df_comeback.head()
```
